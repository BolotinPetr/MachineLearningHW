```python
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, normalized_mutual_info_score
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, VotingClassifier, GradientBoostingRegressor
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
train = pd.DataFrame.from_csv('train.tsv')
#train = train.sample(frac=0.01, random_state=3)

target = train['y']
#train = train.drop(['y'], axis=1)
```

```python
f =  train.columns[4:]
train['mean_f'] = (train[:][f]).mean(axis=1)
```

```python
rf = RandomForestClassifier(n_estimators=100, n_jobs=-1)
rf.fit(train, target)

plt.figure(figsize=(16, 5))
sns.barplot(train.columns, rf.feature_importances_)
_ = plt.xticks(rotation=90)

plt.figure(figsize=(16, 5))
sns.barplot(train.columns, [normalized_mutual_info_score(train[col], target) for col in train.columns])
_ = plt.xticks(rotation=90)
train.groupby('week')['mean_f'].mean().hist()
```
